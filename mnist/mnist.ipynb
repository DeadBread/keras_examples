```python
from keras.models import Sequentialimport keras
```

```python
import kerasfrom keras.datasets import mnist
```

```python
from keras.layers import Dense
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
picts = x_train[:9]
fig = plt.figure()
for i, pict in enumerate(picts):
    fig.add_subplot(3, 3, i + 1)
    plt.imshow(pict, cmap='gray')    
```

```python
model = Sequential()
model.add(Dense(128, activation='relu', input_dim=28*28))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
training_samples = np.zeros([len(y_train), 10])
```

```python
for i, el in enumerate(training_samples):
    el[y_train[i]] = 1
```

```python
test_samples = np.zeros([len(y_test), 10])
for i, el in enumerate(test_samples):
    el[y_test[i]] = 1
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

x_train /= 255
x_test /= 255
```

```python
x_train = x_train.reshape(60000, 28*28)
```

```python
model.fit(x_train, training_samples, epochs=3)
```

```python
model.evaluate(x_test.reshape(10000, 28*28), test_samples)
```
